# Question 5

```python
import numpy as np
import random
import matplotlib.pyplot as plt

alpha = .005
beta = .01
N = 1000
rounds = 80
infected = [0] * N
infected[0] = 1

#LISTS TO KEEP TRACK OF TOTAL INFECTED AT THE END OF EVERY ROUND
total_infected = []

currently_infected_list = [0]

for round in range(0,rounds):
    
    #FOR EACH CURRENTLY INFECTED PERSON
    for person in currently_infected_list:
        
        #ATTEMPT TO INFECT EACH PERSON IN POPULATION
        for i in range(0,N):
            
            #ONLY ATTEMPT TO INFECT IF NOT ALREADY INFECTED
            if i not in currently_infected_list:
                if random.random() < alpha*beta:
                    infected[i] = 1
                    currently_infected_list.append(i)
                    
    total_infected.append(sum(infected))

plt.plot(total_infected)
plt.xlabel("Rounds")
plt.ylabel("Total Infected")
plt.show()
```

# Question 6

```python
import random
import matplotlib.pyplot as plt

alpha = .016
beta = .02
N = 1000
rounds = 100
infected = [-1] * N 
infected[0] = 5

#LISTS TO KEEP TRACK OF STATS AT END OF EVERY ROUND
total_list = [1]
healthy_list = [(N-1)]
sick_list = [1]

#CURRENT STATS
total = 1
healthy = N-1
sick = 1
recovered = 0

currently_infected_indices = [0]

# INFECTED LIST VALUES
#
# -1 = not infected and not immune
# 0  = previously infected, now immune
# 1+ = currently infected with 1+ rounds left of infection 

for round in range(0,rounds):
    
        #FOR EACH CURRENTLY INFECTED PERSON
        for index in currently_infected_indices:

                
            #ATTEMPT TO INFECT THE POPULATION
            for i in range(0,N):
                
                # IF NOT CURRENTLY INFECTED, AND NOT IMMUNE, TRY TO INFECT
                if i not in currently_infected_indices and infected[i] != 0:
                    if random.random() < alpha*beta:
                        infected[i] = 5
                        currently_infected_indices.append(i)
                        total += 1
                        healthy -= 1
                        sick += 1
                        
            #DECREMENT ROUNDS LEFT OF INFECTION
            if infected[index] >= 1:
                infected[index] -= 1
                
            #IF NO LONGER INFECTED, REMOVE FROM CURRENTLY INFECTED
            if infected[index] == 0:
                currently_infected_indices.remove(index)
                healthy += 1
                sick -= 1
                recovered += 1
        
        #UPDATE STATS LISTS
        total_list.append(total)
        healthy_list.append(healthy)
        sick_list.append(sick)
                        
            
plt.plot(total_list)
plt.xlabel("Round")
plt.ylabel("Total Infected")
plt.show()
```

```python
plt.plot(healthy_list)
plt.xlabel("Round")
plt.ylabel("Healthy Individuals")
plt.show()
```

```python
plt.plot(sick_list)
plt.xlabel("Round")
plt.ylabel("Sick Individuals")
plt.show()
```
